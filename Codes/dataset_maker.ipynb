<h1> Resize and Rename IDs </h1>

```python
from PIL import Image

input_path = "E:/University/Term8/Final Project/SOLIDER-REID/merge/"
output_train_path = "E:/University/Term8/Final Project/SOLIDER-REID/iust-v2/IUST/bounding_box_train/"
output_test_path = "E:/University/Term8/Final Project/SOLIDER-REID/iust-v2/IUST/bounding_box_test/"

# Set random seed for reproducibility
random.seed(42)

# Get a list of identity folders (ID0 to ID34)
identity_folders = [f"ID{i}" for i in range(35)]

# Randomly shuffle identity folders
random.shuffle(identity_folders)

# Select 17 identities for training and 18 for testing
train_identities = identity_folders[:17]
test_identities = identity_folders[17:]

# Function to resize images and rename them
def resize_and_rename(input_folder, output_folder, identities):
    for identity in identities:
        identity_folder = os.path.join(input_folder, identity)
        output_identity_folder = os.path.join(output_folder)
        os.makedirs(output_identity_folder, exist_ok=True)
        
        # Get list of images in the identity folder
        images = [img for img in os.listdir(identity_folder) if img.endswith(".jpg")]
        
        # Resize and rename images
        for img in images:
            source_path = os.path.join(identity_folder, img)
            
            # Resize image
            img_data = Image.open(source_path)
            img_data = img_data.resize((64, 128), Image.ANTIALIAS)

            # Adjust image name
            cam_number = img.split('_')[0]
            
            frame_number = img.split('_')[1]
            if len(frame_number) == 5:
                frame_number = frame_number.replace('F','0')
            else:
                frame_number = frame_number.replace('F','')
                
            new_name = f"{int(identity[2:]):04d}_c{cam_number[1]}s1_{frame_number}_00.jpg"
                
            destination_path = os.path.join(output_identity_folder, new_name)
            img_data.save(destination_path)

# Resize and rename images for training
resize_and_rename(input_folder=input_path, output_folder=output_train_path, identities=train_identities)

# Resize and rename images for testing
resize_and_rename(input_folder=input_path, output_folder=output_test_path, identities=test_identities)
```
